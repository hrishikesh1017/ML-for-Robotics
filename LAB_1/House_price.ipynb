Removing Columns Having High Correlation

```python
def correlation(dataset, threshold):
  col_cor = set()
  corr_matrix = dataset.corr()
  for i in range(len(corr_matrix.columns)):
    for j in range(i):
      if abs(corr_matrix.iloc[i,j]) > threshold:
        colname = corr_matrix.columns[i]
        col_cor.add(colname)
  return col_cor
```

```python
def analyze_correlations(df):
    # Ensure columns exist
    if not all(col in df.columns for col in ['bedrooms', 'bathrooms', 'price']):
        raise ValueError("DataFrame must contain 'bedrooms', 'bathrooms', and 'price' columns")

    # Calculate correlation coefficients
    corr_bedrooms_price = df['bedrooms'].corr(df['price'])
    corr_bathrooms_price = df['bathrooms'].corr(df['price'])
    corr_sqft_lot_price = df['floors'].corr(df['floors'])
    corr_sum_price = df[['bedrooms', 'bathrooms']].sum(axis=1).corr(df['price'])
    df['beds_per_floor'] = df['bedrooms'] / df['floors']
    corr_beds_per_sqft_price = df['beds_per_floor'].corr(df['price'])
    print(f"Correlation between beds_per_sqft and price: {corr_beds_per_sqft_price:.4f}")

    # Print correlation results
    print(f"Correlation between bedrooms and price: {corr_bedrooms_price:.4f}")
    print(f"Correlation between bathrooms and price: {corr_bathrooms_price:.4f}")
    print(f"Correlation between floors and price: {corr_sqft_lot_price:.4f}")
    print(f"Correlation between sum of bedrooms and bathrooms and price: {corr_sum_price:.4f}")
    print(f"Correlation between beds_per_floors and price: {corr_beds_per_sqft_price:.4f}")

analyze_correlations(df)
```

Visualizing and removing outliers using boxplots

```python
def remove_outliers_using_boxplot(df, column):

    # Visualize the boxplot
    plt.figure(figsize=(10, 6))
    plt.boxplot(df[column], vert=False)
    plt.title(f'Boxplot of {column}')
    plt.xlabel(column)
    plt.show()
    
    # Calculate the IQR
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    # Determine the lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Remove outliers
    df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    
    return df_filtered
```

Train Test Splitting

```python
X = df.iloc[:,:-1].values
Y = df.iloc[:,-1].values
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.1,random_state=42)
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train[:,0:] = sc.fit_transform(x_train[:,0:])
x_test[:,0:] = sc.transform(x_test[:,0:])
```

Training Using Linear Model

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
```

Training Using RandomForest Model

```python
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor()
rf_model.fit(x_train, y_train)
```

Training using SVM model

```python
from sklearn.svm import SVR
svm_model = SVR(kernel='linear')
svm_model.fit(x_train, y_train)
```

Evaluation Function

```python
from sklearn.metrics import r2_score, mean_squared_error, root_mean_squared_error

def evaluate(model, y_test, x_test):
    y_pred = model.predict(x_test)
    r2 = r2_score(y_test, y_pred)
    mse = mean_squared_error(y_test,y_pred)
    rmse = root_mean_squared_error(y_test,y_pred)
    print("r2 score = ",r2)
    print("mean_squared_error = ",mse)
    print("root_mean_squared_error = ",rmse)
```

Visualization Function

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_prediction_vs_actual(y_test, y_pred):


    plt.figure(figsize=(8, 6))

    # Scatter plot
    plt.scatter(y_test, y_pred, color='blue', label='Predicted vs Actual')
    
    # Add 45-degree line
    min_val = min(min(y_test), min(y_pred))
    max_val = max(max(y_test), max(y_pred))
    plt.plot([min_val, max_val], [min_val, max_val], color='red', linestyle='--', label='Perfect Fit')

    # Labels and title
    plt.xlabel('Actual Values')
    plt.ylabel('Predicted Values')
    plt.title('Prediction vs Actual')
    plt.legend()
    
    # Show plot
    plt.grid(True)
    plt.show()

```

EVALUATIONS AND VISUALIZATIONS OF MODELS

```python
#LINEAR MODEL

evaluate(linear_model,y_test=y_test, x_test=x_test)

# RANDOM FOREST MODEL

evaluate(rf_model,y_test=y_test, x_test=x_test)

#SUPPORT VECTOR MACHINE

evaluate(svm_model,y_test=y_test, x_test=x_test)
```
